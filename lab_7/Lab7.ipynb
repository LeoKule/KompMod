# Лабораторная работа №7. "Модели конкуренции"
## Вариант 17

Межвидовая крнкуренция. В этом случае исследуется конкуренция популяций, потребляющих общий ресурс. Пусть $N_1$ и $N_2$ - численности конкурирующих популяций. Модель (Лотки-Вольтерры) выражается уровнениями:
$$
\begin{cases}
    \frac{dN_1}{dt} = r_1N_1\frac{K_1-N_1-\alpha_{12}N_2}{K_1} \\
    \frac{dN_2}{dt} = r_2N_2\frac{K_2-N_2-\alpha_{21}N_1}{K_2}
\end{cases}
$$

Параметры $a_{12}$ и $a_{21}$ отражают итенсивность межвидовой конкуренции. <br> 
Главный вопрос, который интересует исследователя межвидовой конкуренции, - при каких условиях увеличивается или уменьшается численность каждого вида? <br>
Данная модель предсказывает следующие режимы эволюции взаимодествующих популяций: устойчивое сосуществование или полное вытеснение одной из них. <br>
Построить в фазовой плоскости $(N_1, N_2)$ границы зон, разделяющих какие-либо два режима эволюции конкурирующих популяций. Остальные параметры модели выбрать произвольно. Учесть при этом, что режим устойчивого сосуществования популяции может в принипе реализоваться только при $a_{12}<1$, $a_{21}<1$. <br>
Визуализировать в виде графиков зависимости от $t$ и фазовых диаграмм, на котрых указать особые точки, изоклины и направления на траекториях. 

## Решение 
Все коэффициенты неотридцательные <br><br>


#### Найдём особые точки:
$$
\begin{cases}
    r_1N_1 \frac{K_1-N_1-\alpha_{12}N_2}{K_1} = 0 \\
    r_2N_2 \frac{K_2-N_2-\alpha_{21}N_1}{K_2} = 0
\end{cases}
$$


$$
\begin{cases}
    N_1(r_1K_1-r_1N_1-\alpha_{12}r_1N_2) = 0 \\ 
    N_2(r_2K_2-r_2N_2-\alpha_{21}r_2N_1) = 0
\end{cases}
$$

Особая точка №1
$$
\begin{align*}
    N_1^{(1)} = 0 \\
    N_2^{(1)} = 0
\end{align*}
$$

Особая точка №2
\begin{align*}
    N_1^{(2)} = 0 \\
    r_2K_2-r_2N_2 =0
\end{align*}


\begin{align*}
    N_1^{(2)} &= 0 \\
    N_2^{(2)} &= K_2
\end{align*}

Особая точка №3


\begin{align*}
    r_1K_1-r_1N_1 = 0 \\
    N_2^{(3)} = 0
\end{align*}



\begin{align*}
    N_1^{(3)} &= K_1 \\
    N_2^{(3)} &= 0
\end{align*}


Особая точка №4

\begin{cases}
    K_1-N_1-\alpha_{12}N_2 = 0 \\ 
    K_2-N_2-\alpha_{21}N_1 = 0
\end{cases}



\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    K_2-N_2-\alpha_{21}(K_1 - \alpha_{12}N_2 ) = 0
\end{cases}



\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    K_2-N_2-\alpha_{21}K_1 + \alpha_{21}\alpha_{12}N_2 = 0
\end{cases}




\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    N_2 = \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1}
\end{cases}



\begin{cases}
    N_1^{(4)} =  K_1 - \frac{\alpha_{12}\alpha_{21}K_1 - \alpha_{12}K_2}{\alpha_{21}\alpha_{12} - 1} =  \frac{\alpha_{12}K_2 -K_1}{\alpha_{21}\alpha_{12} - 1} \\ 
    N_2^{(4)} = \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1}
\end{cases}


#### Определим фазовый портрет 
$$J = 
\begin{pmatrix}\frac{r_1(K_1 - 2N_1-N_2\alpha_{12})}{K_1} & -\frac{r_1\alpha_{12}N_1}{K_1} \\ -\frac{r_2\alpha_{21}N_2}{K_2} & \frac{r_2(K_2 - 2N_2-N_1\alpha_{21})}{K_2}\end{pmatrix}
$$



Особая точка $(0,0)$
$$
\begin{bmatrix}
    r_1  - \lambda & 0 \\
    0 & r_2  - \lambda 
\end{bmatrix}
=
(r_1  - \lambda)(r_2 - \lambda ) = 0
\implies
\begin{cases}
    \lambda_{1} = r_1 \\
    \lambda_{2} = r_2
\end{cases}
\implies
неустойчивый \quad узел
$$

Особая точка $(0, K_2)$
$$
\det \begin{bmatrix}
    \frac{r_1K_1 - r_1K_2\alpha_{12}}{K_1} - \lambda & 0 \\
    -r_2\alpha_{21} & \frac{r_2K_2 - 2r_2K_2}{K_2} - \lambda 
\end{bmatrix}
= \left(\frac{r_1K_1 - r_1K_2\alpha_{12}}{K_1} - \lambda\right)
  \left(\frac{r_2K_2 - 2r_2K_2}{K_2} - \lambda\right) = 0 
$$
\begin{cases}
    \lambda_{1} = r_1 - \frac{r_1K_2\alpha_{12}}{K_1} \\
    \lambda_{2} = -r_2
\end{cases}

\begin{cases}
    устойчивый \quad узел, \quad K_1 < K_2\alpha_{12} \\
    седло, \quad иначе
\end{cases}


Особая точка $(K_1, 0)$

$$
\det \begin{bmatrix}
    \frac{r_1K_1 - 2r_1K_1}{K_1} - \lambda & -r_1\alpha_{12} \\
    0 & \frac{r_2K_2 - r_2K_1\alpha_{21}}{K_2} - \lambda
\end{bmatrix}
=
\left(\frac{r_1K_1 - 2r_1K_1}{K_1} - \lambda\right)
\left(\frac{r_2K_2 - r_2K_1\alpha_{21}}{K_2} - \lambda\right)
= 0
$$


\begin{cases}
    \lambda_{1} = -r_1  \\
    \lambda_{2} = r_2  -\frac{r_2K_1\alpha_{21}}{K_2}
\end{cases}

\begin{cases}
    \text{устойчивый узел}, \quad K_2 < K_1\alpha_{21} \\
    \text{седло}, \quad \text{иначе}
\end{cases}

Особая точка $( \frac{\alpha_{12}K_2 -K_1}{\alpha_{21}\alpha_{12} - 1}, \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1})$


$
\begin{bmatrix}
   -\frac{r_1}{K_1}N_1 - \lambda  & -\frac{r_1\alpha_{12}}{K_1}N_1 \\
   -\frac{r_2\alpha_{21}}{K_2}N_2 & -\frac{r_2}{K_2}N_2-\lambda 
\end{bmatrix}= 
 (-\frac{r_1}{K_1}N_1 - \lambda)(-\frac{r_2}{K_2}N_2-\lambda ) -\frac{r_1r_2\alpha_{12}\alpha_{21}}{K_1K_2}N_1N_2 = \lambda^2 + \lambda(\frac{r_1}{K_1}N_1+\frac{r_2}{K_2}N_2) + \frac{r_1r_2}{K_1K_2}N_1N_2 -\frac{r_1r_2\alpha_{12}\alpha_{21}}{K_1K_2}N_1N_2 = 0
$

$$
\lambda^2 + \lambda\frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2} + \frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2}  = 0
$$

$$
\lambda_{1}\lambda_{2} = \frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2}
$$
$$\frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2} > 0$$
$$r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21}) > 0$$
$$\frac{(\alpha_{12}K_2 -K_1)(\alpha_{21}K_1 - K_2)(1-\alpha_{12}\alpha_{21})}{(\alpha_{21}\alpha_{12} - 1)^2} > 0$$
$$(\alpha_{12}K_2 -K_1)(\alpha_{21}K_1 - K_2) > 0$$
$$\alpha_{21}\alpha_{12}K_1K_2-\alpha_{12}K_2^2-\alpha_{21}K_1^2+K_1K_2 >0$$
$$K_1K_2 > \alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2$$
$$K_1 > \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2}$$
От сюда следует
$$
\begin{cases}
    узел \quad при \quad K_1 > \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2} \\
    седло \quad иначе
\end{cases}
$$

$$
\lambda_{1} + \lambda_{2} = - \frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2}
$$

$$
- \frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2} < 0
$$

$$ \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2} < K_1 \quad - \quad устойчивый \quad узел$$

#### Определим изоклины
$$\begin{cases}
    r_1N_1 \frac{K_1-N_1-\alpha_{12}N_2}{K_1} = 0\\
    r_2N_2 \frac{K_2-N_2-\alpha_{21}N_1}{K_2} = 0
\end{cases}$$
Уравнение изоклин горизонтальных касательных 
$$ N_1 = 0 \quad и \quad N_2 =\frac{K_1-N_1}{\alpha_{12}} $$
Уравнение изоклин вертикальных касательных
$$N_2 = 0 \quad и \quad  N_2= K_2-\alpha_{21}N_1$$

```python
# Библиотеки 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Модель Лотки-Вольтерры
def competition_model(N, t, r1, r2, K1, K2, alpha12, alpha21):
    N1, N2 = N
    dN1_dt = r1 * N1 * (K1 - N1 - alpha12 * N2) / K1
    dN2_dt = r2 * N2 * (K2 - N2 - alpha21 * N1) / K2
    return [dN1_dt, dN2_dt] 

# Фазовый портрет
def plot_phase_portrait(ax, r1, r2, K1, K2, alpha12, alpha21, title):
    points = []    # Особых точки
    point_phase = []
    points.append((0, 0))  
    point_phase.append('Неустойчивый узел')
    points.append((K1, 0))  
    point_phase.append('Устойчивый узел' if K2 < alpha21*K1 else 'Седло')
    points.append((0, K2))  
    point_phase.append('Устойчивый узел' if K1 < alpha12*K2 else 'Седло')
    denom = 1 - alpha12 * alpha21
    if abs(denom) > 1e-10:
        N1_star = (K1 - alpha12 * K2) / denom
        N2_star = (K2 - alpha21 * K1) / denom
        if N1_star > 0 and N2_star > 0:
            points.append((N1_star, N2_star))
            point_phase.append('Устойчивый узел' if K1 > (alpha21*(K1**2)+alpha12*(K2**2)-alpha12*alpha21*K1*K2)/K2 else 'Седло')
    
    # Создание сетки для фазового портрета
    N1 = np.linspace(0, K1 * 1.2, 20)
    N2 = np.linspace(0, K2 * 1.2, 20)
    N1_grid, N2_grid = np.meshgrid(N1, N2)
    
    # Вычисление скоростей
    dN1_dt = np.zeros_like(N1_grid)
    dN2_dt = np.zeros_like(N2_grid)
    
    for i in range(len(N1)):
        for j in range(len(N2)):
            dN = competition_model([N1_grid[i,j], N2_grid[i,j]], 0, 
                                 r1, r2, K1, K2, alpha12, alpha21)
            dN1_dt[i,j] = dN[0]
            dN2_dt[i,j] = dN[1]
    
    # Нормализация стрелок
    magnitude = np.sqrt(dN1_dt**2 + dN2_dt**2)
    dN1_dt_norm = np.where(magnitude > 0, dN1_dt/magnitude, 0)
    dN2_dt_norm = np.where(magnitude > 0, dN2_dt/magnitude, 0)
    
    # Построение векторного поля
    ax.quiver(N1_grid, N2_grid, dN1_dt_norm, dN2_dt_norm, 
              color='blue', alpha=0.6, scale=30)
    
    # Изоклины
    N1_iso = np.linspace(0, K1 * 1.2, 100)
    # dN1/dt = 0: N2 = (K1 - N1)/alpha12
    N2_iso1 = (K1 - N1_iso) / alpha12
    # dN2/dt = 0: N2 = K2 - alpha21*N1
    N2_iso2 = K2 - alpha21 * N1_iso
    
    ax.plot(N1_iso, N2_iso1, 'black', linewidth=1)
    ax.plot(N1_iso, N2_iso2, 'black', linewidth=1)
    ax.axvline(x=0, ymin=0, ymax=K2*1.2, color='black', linewidth=1)  
    ax.plot([0, K1*1.2], [0, 0], 'black', linewidth=1)
    
    # Особые точки
    color = ['yellow', 'green', 'blue', 'purple']
    for i, (x, y) in enumerate(points):
        if x >= 0 and y >= 0 and x <= K1*1.2 and y <= K2*1.2:
            ax.plot(x, y, 'o', markersize=8, color=color[i], label=point_phase[i])
    
    # Траектории из разных начальных условий
    initial_conditions = [
        [0.1, 0.1], [0.5, 0.1], [0.1, 0.5],
        [0.8, 0.2], [0.2, 0.8], [0.6, 0.6]
    ]
    
    t = np.linspace(0, 20, 1000)
    for N0 in initial_conditions:
        solution = odeint(competition_model, N0, t, 
                         args=(r1, r2, K1, K2, alpha12, alpha21))
        ax.plot(solution[:, 0], solution[:, 1], 'red', alpha=0.7, linewidth=1)
        ax.plot(solution[0, 0], solution[0, 1], 'ko', markersize=3)
    
    ax.set_xlabel('N₁ (Популяция вида 1)')
    ax.set_ylabel('N₂ (Популяция вида 2)')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-0.02, K1 * 1.2)
    ax.set_ylim(-0.05, K2 * 1.2)

def plot_time_series(ax, r1, r2, K1, K2, alpha12, alpha21, title):
    """Построение временных рядов"""
    t = np.linspace(0, 20, 1000)
    
    # Разные начальные условия
    initial_conditions = [[0.5, 0.5], [0.8, 0.2]]
    colors = ['red', 'blue']
    
    for i, N0 in enumerate(initial_conditions):
        solution = odeint(competition_model, N0, t, 
                         args=(r1, r2, K1, K2, alpha12, alpha21))
        
        ax.plot(t, solution[:, 0], color=colors[0], 
               linestyle=['-.', '-'][i])
        ax.plot(t, solution[:, 1], color=colors[1], 
               linestyle=['--', '-'][i])
    
    ax.set_xlabel('Время')
    ax.set_ylabel('Численность')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    
params = {
    "Сосуществование": {
        "r1": 1.0, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 0.5, "alpha21": 0.5
    },
    "Победа вида 1": {
        "r1": 1.0, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 0.5, "alpha21": 1.5
    },
    "Победа вида 2": {
        "r1": 1.0, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 1.5, "alpha21": 0.5
    },
    "Бистабильность": {
        "r1": 1.0, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 1.2, "alpha21": 1.2
    }
}
   
fig, axes = plt.subplots(4, 2, figsize=(15, 20))

for i, (regime, param) in enumerate(params.items()):
    # Фазовый портрет
    plot_phase_portrait(axes[i,0], 
                       param["r1"], param["r2"], 
                       param["K1"], param["K2"],
                       param["alpha12"], param["alpha21"],
                       f"Фазовый портрет: {regime}")
    plot_time_series(axes[i, 1],
                    param["r1"], param["r2"], 
                    param["K1"], param["K2"],
                    param["alpha12"], param["alpha21"],
                    f"Временные ряды: {regime}")
```
